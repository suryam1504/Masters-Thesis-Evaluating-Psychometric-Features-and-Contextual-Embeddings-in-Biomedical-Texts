```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
df_larger_annotated = pd.read_csv('/content/drive/MyDrive/IDSIA Biomedical Texts/Tweets Project/large16k_with_annotated_part_and_feat_incl_emb.csv')
df_larger_annotated
```

```python
# converting embeddings from string to list
df_larger_annotated['twitter4sse_embeddings'] = df_larger_annotated['twitter4sse_embeddings'].apply(lambda x: eval(x))
```

```python
df_annotated = df_larger_annotated[(df_larger_annotated.used_for_annotation==1) & (df_larger_annotated.majority_vote!="NoMajority")]
df_annotated
```

```python
df_annotated[df_annotated['majority_vote'] == 'PP']['twitter4sse_embeddings']
```

```python
pp_avg_embedding = np.mean(np.vstack(df_annotated[df_annotated['majority_vote'] == 'PP']['twitter4sse_embeddings'].values), axis=0)
pp_avg_embedding
```

```python
len(pp_avg_embedding)
```

```python
ponp_avg_embedding = np.mean(np.vstack(df_annotated[df_annotated['majority_vote'] == 'PONP']['twitter4sse_embeddings'].values), axis=0)
```

```python
un_avg_embedding = np.mean(np.vstack(df_annotated[df_annotated['majority_vote'] == 'UN']['twitter4sse_embeddings'].values), axis=0)
```

```python
df_messy = df_larger_annotated[(df_larger_annotated.used_for_annotation==1) & (df_larger_annotated.majority_vote=="NoMajority")]
df_messy.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
type(df_messy['twitter4sse_embeddings'][381])
```

```python
cosine_similarity([df_messy['twitter4sse_embeddings'][381]], [list(pp_avg_embedding)])
```

```python
cosine_similarity([df_messy['twitter4sse_embeddings'][381]], [list(ponp_avg_embedding)])
```

```python
cosine_similarity([df_messy['twitter4sse_embeddings'][381]], [list(un_avg_embedding)])
```

```python
for i in list(df_messy.index):
  print(i, cosine_similarity([df_messy['twitter4sse_embeddings'][i]], [list(pp_avg_embedding)]))
  print(i, cosine_similarity([df_messy['twitter4sse_embeddings'][i]], [list(ponp_avg_embedding)]))
  print(i, cosine_similarity([df_messy['twitter4sse_embeddings'][i]], [list(un_avg_embedding)]))
  print()
```
