```python
from transformers import RobertaTokenizerFast, TFRobertaForSequenceClassification, pipeline

tokenizer = RobertaTokenizerFast.from_pretrained("arpanghoshal/EmoRoBERTa")
model = TFRobertaForSequenceClassification.from_pretrained("arpanghoshal/EmoRoBERTa")

#emotion = pipeline('sentiment-analysis', model='arpanghoshal/EmoRoBERTa')
emotion = pipeline('sentiment-analysis', model='arpanghoshal/EmoRoBERTa' , return_all_scores= True)  # https://stackoverflow.com/questions/71557275/emotion-detection-in-text-using-emoroberta
```

```python
# Assuming you have the predicted probabilities for each emotion stored in `emotion_probabilities`

# Define the intensity scale for each emotion
intensity_scale = {
    'admiration': (1, 3),  # Example: Mapping fear to a 3-scale (1: no fear, 2: somewhat fear, 3: extreme fear)
    'amusement': (1, 5),
    'anger': (1, 3),
    'annoyance': (1, 3),
    'approval': (1, 3),
    'caring': (1, 3),
    'confusion': (1, 3),
    'curiosity': (1, 3),
    'desire': (1, 3),
    'disappointment': (1, 3),
    'disapproval': (1, 3),
    'disgust': (1, 3),
    'embarrassment': (1, 3),
    'excitement': (1, 3),
    'fear': (1, 3),
    'gratitude': (1, 3),
    'grief': (1, 3),
    'joy': (1,3),
    'love': (1, 3),
    'nervousness': (1, 3),
    'optimism': (1, 3),
    'pride': (1,3),
    'realization': (1, 3),
    'relief': (1, 3),
    'remorse': (1, 3),
    'sadness': (1, 3),
    'surprise': (1, 3),
    'neutral': (1, 3)  # Example: Mapping joy to a 5-scale (1: no joy, 2-4: varying levels of joy, 5: extreme joy)
    # Define scales for other emotions as per your requirement
}

intensity_values = {}

# Assign intensities for each emotion based on probabilities
for emotion, probs in {d['label']: d['score'] for d in emotion_labels[0]}.items():
    min_intensity, max_intensity = intensity_scale[emotion]
    intensity = min_intensity + (max_intensity - min_intensity) * probs
    intensity = round(intensity)  # Round intensity to the nearest whole number if necessary
    intensity_values[emotion] = intensity

# intensity_values dictionary will contain the assigned intensities for each emotion in the desired scale
```
