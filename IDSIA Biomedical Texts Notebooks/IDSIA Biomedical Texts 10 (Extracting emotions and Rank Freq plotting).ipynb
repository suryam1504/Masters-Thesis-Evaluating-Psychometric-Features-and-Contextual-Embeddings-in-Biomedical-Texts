```python
import tensorflow as tf
```

```python
from transformers import RobertaTokenizerFast, TFRobertaForSequenceClassification
```

```python
tokenizer = RobertaTokenizerFast.from_pretrained("arpanghoshal/EmoRoBERTa")
model = TFRobertaForSequenceClassification.from_pretrained("arpanghoshal/EmoRoBERTa")
```

```python
text = "this is too much I cannot handle it"
encoded_input = tokenizer(text, return_tensors="tf")
output = model(encoded_input["input_ids"], attention_mask=encoded_input["attention_mask"])

emotions = output.logits[0]#.numpy() # these are logits, not probabilities, this needs torch.sigmoid() or torch.nn.functional.softmax() (Ref- https://stackoverflow.com/questions/46416984/how-to-convert-logits-to-probability-in-binary-classification-in-tensorflow)
emotions_prob = tf.math.softmax(emotions).numpy() # we use tensorflow softmax here to convert logits to probabilities

emotion_labels = model.config.id2label.values()
emotion_probabilities = dict(zip(emotion_labels, emotions_prob))
```

```python
emo_label_prob = []  # took 26 mins to run with GPU runtime

for item in text_list:
  encoded_input = tokenizer(item, truncation=True, padding=True, return_tensors="tf")
  output = model(encoded_input["input_ids"], attention_mask=encoded_input["attention_mask"])

  emotions = output.logits[0]#.numpy() # these are logits, not probabilities, this needs torch.sigmoid() or torch.nn.functional.softmax() (Ref- https://stackoverflow.com/questions/46416984/how-to-convert-logits-to-probability-in-binary-classification-in-tensorflow)
  emotions_prob = tf.math.softmax(emotions).numpy() # we use tensorflow softmax here to convert logits to probabilities

  emotion_labels = model.config.id2label.values()
  emotion_probabilities = dict(zip(emotion_labels, emotions_prob))

  emo_label_prob.append(emotion_probabilities)
```

```python
# code so that values of common keys in dictionary add up in average (using collections.Counter())
from collections import Counter

dict_list = [{'A': 1, 'B': 2, 'C': 3},
             {'A': 10, 'B': 2, 'D': 3},
             {'A': 5, 'B': 1, 'C': 2}]

dict_list = [my_dict, my_dict1, my_dict2, my_dict3]

merged_counter = Counter()
counts = Counter()

for dictionary in dict_list:
    merged_counter.update(dictionary)
    counts.update(dictionary.keys())

merged_dict = {key: merged_counter[key] / counts[key] for key in merged_counter}

print(merged_dict)
```

```python
df['emo_label_rank'] = df['emo_labels'].apply( lambda x: {key: value for key, value in zip(x, [1,2,3,4,5])} )
```

```python
df.head(2)
```

```python
### MAIN CODE

# code so that values of common keys in dictionary add up in average (using collections.Counter())
from collections import Counter

dict_list = list(df['emo_label_rank'])

merged_counter = Counter()
counts = Counter()

for dictionary in dict_list:
    merged_counter.update(dictionary)
    counts.update(dictionary.keys())

merged_dict_avg = {key: merged_counter[key] / counts[key] for key in merged_counter}

print(merged_dict_avg)
```

```python
# to check if its giving correct average

# code so that values of common keys in dictionary add up directly
dict_list = list(df['emo_label_rank'])

merged_dict_add = {}

for dictionary in dict_list:
    for key, value in dictionary.items():
        merged_dict_add[key] = value + merged_dict_add.get(key, 0)

print(merged_dict_add)
```
