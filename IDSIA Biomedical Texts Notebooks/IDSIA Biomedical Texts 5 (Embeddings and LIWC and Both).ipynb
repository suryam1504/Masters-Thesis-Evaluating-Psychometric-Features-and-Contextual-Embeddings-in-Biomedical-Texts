```python
# Logistic Regression
clf = LogisticRegression(solver='lbfgs', max_iter=11000)
f1_list = []
roc_auc_list = []
for _ in range(10):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None, stratify = y)
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  f1 = f1_score(y_test, y_pred,average='weighted')
  f1_list.append(f1)
  rocaucscore = roc_auc_score(y_test, y_pred)
  roc_auc_list.append(rocaucscore)
print('10 f1 scores list: ', f1_list)
print('10 roc auc scores list: ', roc_auc_list)
print('Mean of f1 scores:', np.mean(f1_list))
print('Standard Deviation of f1 scores', np.std(f1_list))
print('Mean of ROC AUC scores:', np.mean(roc_auc_list))
print('Standard Deviation of ROC AUC scores', np.std(roc_auc_list))
```
