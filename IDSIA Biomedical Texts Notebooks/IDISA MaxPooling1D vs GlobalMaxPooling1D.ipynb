```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, GlobalMaxPooling1D
```

```python
# Example input tensor
input_tensor = Input(shape=(26, 1))  # 100 timesteps, 10 features

# Apply MaxPooling1D with pool_size=2 and strides=2
max_pooling_output = MaxPooling1D(pool_size=3, strides=1)(input_tensor)

# Apply GlobalMaxPooling1D
global_max_pooling_output = GlobalMaxPooling1D()(input_tensor)

# Print output shapes
print("MaxPooling1D output shape:", max_pooling_output.shape)
print("GlobalMaxPooling1D output shape:", global_max_pooling_output.shape)
```
