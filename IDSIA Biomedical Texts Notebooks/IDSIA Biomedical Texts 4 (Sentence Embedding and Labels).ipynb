## Getting Word Embeddings using Sentence Transformers

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
# https://sbert.net/ - for more models, commonly used ones are 'bert-base-nli-mean-tokens' and 'distilbert-base-nli-mean-tokens'
```

```python
# Dataset of texts
sentences = list(df['text'])
```

```python
sentence_embeddings = model.encode(sentences)  # 45 sec to run with gpu
```

## Preparing X and y for Sentence Embbedings and Labels

```python
from sklearn.model_selection import train_test_split
```

```python
# not label encoding the 'label' column is still fine
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])
df.head() # gives 1 for panic, 0 for anxiety
```

```python
X = sentence_embeddings
y = df['label']
```

```python
sum(y)/len(y)  # might wanna use stratification
```

```python
len(X_train), len(y_train)
```

```python
clf.fit(X_train, y_train) 
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# F1-score = 2*(Recall*Precision)/Recall+Precision where,

# Precision = TP/TP+FP
# Recall = TP/TP+FN
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
print('precision: %.2f' % precision_score(y_test, y_pred,average='weighted'))
print('recall: %.2f' % recall_score(y_test, y_pred,average='weighted'))
print('f1_score: %.2f' % f1_score(y_test, y_pred,average='weighted'))
```
